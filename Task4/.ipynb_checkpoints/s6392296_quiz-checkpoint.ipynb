```python
# Quiz 4.14

import numpy as np

def funcfilter(func,seq):
    b = []
    seq = np.array(seq,ndmin=1)
    
    for s in seq:
        if s > 0:
            b.append(s)
    
    print(b)
    f = func(b)
    return f

a = np.array([-100, -10, 0, 10, 100])
print(funcfilter(np.log10,a))   
        
print(funcfilter(np.log10,100))  
```

```python
# Quiz 4.16

from astropy.cosmology import Planck18
from scipy import integrate
import numpy as np

H0_km = Planck18.H0.value
Om0 = Planck18.Om0
Ode0 = Planck18.Ode0
H0 = H0_km*1000/3.08568


def F(z):
    H = H0*((Om0*(1+z)**3)+Ode0)**0.5
    t = 1/((1+z)*H)
    return t

a = 0
z = 2

integ, error = integrate.quad(F, a, z) #does the integral
time = (integ/(365*24*60*60))*10e12
print("t(z) =", time, "Gyr")

```

```python
# Quiz 4.21
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import norm

x = np.random.normal(0, 1, 100000)
y = np.random.normal(2, 1, 100000)
xp =np.linspace(min(x),max(x),100000)
yp =np.linspace(min(y),max(y),100000)
plt.hist(x, bins=30, alpha=0.6, color='blue', density=True, label="x", edgecolor='black')
plt.hist(y, bins=30, alpha=0.6, color='violet',density=True, label="y", edgecolor='black')
plt.plot(xp, norm.pdf(xp,0,1), color='black', linestyle='dotted', label="X PDF")
plt.plot(yp, norm.pdf(yp,2,1), color='black', linestyle='dashed', label="Y PDF")
plt.legend()
plt.xlabel("Values")
plt.ylabel("Probability Density")
plt.annotate('N=(0,1)', xy=(3.5, 0.2), xytext=(4.8, 0.2),
            arrowprops=dict(color='violet', shrink=0.05))

plt.annotate('N=(0,1)', xytext=(-2.8, 0.2), xy=(-1.5, 0.2) , ha='right',
            arrowprops=dict(color='blue', shrink=0.05))

plt.title("1D Hystogram + 1D")
plt.grid(axis = 'y', linestyle='dashed')
```

```python
import numpy as np
from matplotlib import pyplot as plt
from scipy.stats import norm


x = np.random.normal(0, 1, 100000)
y = np.random.normal(2, 1, 100000)

h, xe, ye, _ = plt.hist2d(x, y, bins=30, cmap='Blues')

nr = h/h.sum()
mr = nr.flatten()
pr = np.sort(mr)[::-1]
summ = np.cumsum(pr)

levels = [pr[np.searchsorted(summ, a)] for a in [0.68,0.95,0.997]]
levels = sorted(levels)

Xcen = 0.5 * (xe[1:] + xe[:-1])
Ycen = 0.5 * (ye[1:] + ye[:-1])
X, Y = np.meshgrid(Xcen, Ycen)

plt.gca().set_aspect(1)
plt.colorbar()
plt.grid(linestyle='dashed', color='black')
plt.ylabel("Y")
plt.xlabel("X")
plt.title("2D Hystogram + Contour")
CS = plt.contour(X, Y, nr.T,levels=levels, colors='red')
plt.clabel(CS, fmt={levels[2]: '1std', levels[1]: '2std', levels[0]: '3std'}, inline=True)
```
