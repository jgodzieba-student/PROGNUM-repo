```python
# Quiz 3.2
from scipy.constants import G

masses = [1.9891e+30, 1.8986e+27, 5.6846e+26, 
          1.0243e+26, 8.6810e+25, 5.9736e+24, 
          4.8685e+24, 6.4185e+23, 3.3022e+23, 
          7.349e+22, 1.25e22]
radii  = [696340, 71492, 60268, 24764, 
          25559, 6378.1, 6051.8, 3396.2, 
          2439.7, 1738.1, 1195]
names  = ['Sun', 'Jupiter', 'Saturn', 'Neptune', 
          'Uranus', 'Earth', 'Venus', 'Mars', 
          'Mercury', 'Moon', 'Pluto']
solarv = {}
for m in masses:
    v = 2*G*m
    print(f"mass is {m:.2f}")
for r in radii:
    velocity = (v/r)**0.5
for n in names:
    solarv[n] = velocity
print(solarv)
```

```python
# Quiz 3.17

import numpy as np

def stdev(A):
    N = len(A)
    y = sum(A)
    m = y/N
    s = 0
    i = 0
    for x in A:
        while i < (N-1):
            s += (x-m)**2
            i += 1
        std = ((1/N)*s)**0.5
        return std

print(stdev([1,2,3,4]))

print(np.std([1,2,3,4]))

def stdev(A):
    s = 0
    i = 0
    for x in A:
        while i < (len(A)-1):
            s += (x-(sum(A)/len(A)))**2
            i += 1
        std = ((1/len(A))*s)**0.5
        return std

print(stdev([1,2,3,4]))
```
